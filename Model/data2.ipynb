```python
def one_hot_encoding(df, columns):
    for c in columns:
        one_hot = pd.get_dummies(df[c], prefix = c)
        df = pd.concat([df.drop(c, axis = 1), one_hot], axis = 1)
    return df

final_same = same.copy()
final_int = interp.copy()

columns_encode = ['US_President', 'KR_President']
```

# 3 아웃라이어 전 + standardi

```python
log_cols = ['Open', 'High', 'Low', 'Close', 'Volume', 'Transaction', 'SE_Operating_Profit', 'SE_Net_Profit', 'SE_Total_Equity', 'SE_Total_Liabilities', 'SE_Revenue', 'SE_Foreign', 'SOX_Close', 'Phone']
standardize_cols = ['Change', 'US_Inflation', 'KR_Inflation', 'CN_Inflation', 'US_GDP_Annual', 'KR_GDP_Annual', 'CN_GDP_Annual', 'US_Economic_Growth_Rate', 'KR_Economic_Growth_Rate', 'US_Base', 'US_10', 'KR_Base', 'KR_10', 'WTI', 'Exchange', 'KR_Unemployment']
```

```python
from sklearn.preprocessing import StandardScaler

def log_scale_transform(data, log_cols, standardize_cols):
    log_transformed_data = np.log(data[log_cols].replace(0, 1e-10))
    log = pd.DataFrame(log_transformed_data, columns=log_cols, index=data.index)

    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(data[standardize_cols])
    scaled = pd.DataFrame(scaled_data, columns=standardize_cols, index=data.index)

    unchanged_cols = data.columns.difference(log_cols + standardize_cols)
    unchanged_df = data[unchanged_cols]

    log_scale = pd.concat([log, scaled, unchanged_df], axis=1)

    log_scale = log_scale[data.columns]

    return log_scale
```

```python
same_log_scale = log_scale_transform(same, log_cols, standardize_cols)
same_log_scale.head()
```

```python
int_log_scale = log_scale_transform(interp, log_cols, standardize_cols)
int_log_scale.head()
```

```python
def one_hot_encoding(df, columns):
    for c in columns:
        one_hot = pd.get_dummies(df[c], prefix = c)
        df = pd.concat([df.drop(c, axis = 1), one_hot], axis = 1)
    return df

final_same = same_log_scale.copy()
final_int = int_log_scale.copy()

columns_encode = ['US_President', 'KR_President']
```

```python
###Same
final_same_df = one_hot_encoding(final_same, columns_encode)
final_same_df = final_same_df.rename(columns = {'US_President_1.0' : 'US_President_Republican',
       'US_President_2.0' : 'US_President_Democratic', 'KR_President_1.0' : 'KR_President_Democratic',
       'KR_President_2.0' : 'KR_President_Conservative','KR_President_3.0' : 'KR_President_Neutral'})

final_same_df['US_President_Republican'] = final_same_df['US_President_Republican']
final_same_df['US_President_Democratic'] = final_same_df['US_President_Democratic']
final_same_df['KR_President_Democratic'] = final_same_df['KR_President_Democratic']
final_same_df['KR_President_Conservative'] = final_same_df['KR_President_Conservative']
final_same_df['KR_President_Neutral'] = final_same_df['KR_President_Neutral']


###Interp
final_int_df = one_hot_encoding(final_int, columns_encode)
final_int_df = final_int_df.rename(columns = {'US_President_1.0' : 'US_President_Republican',
       'US_President_2.0' : 'US_President_Democratic', 'KR_President_1.0' : 'KR_President_Democratic',
       'KR_President_2.0' : 'KR_President_Conservative','KR_President_3.0' : 'KR_President_Neutral'})
final_int_df['US_President_Republican'] = final_int_df['US_President_Republican']
final_int_df['US_President_Democratic'] = final_int_df['US_President_Democratic']
final_int_df['KR_President_Democratic'] = final_int_df['KR_President_Democratic']
final_int_df['KR_President_Conservative'] = final_int_df['KR_President_Conservative']
final_int_df['KR_President_Neutral'] = final_int_df['KR_President_Neutral']
```
