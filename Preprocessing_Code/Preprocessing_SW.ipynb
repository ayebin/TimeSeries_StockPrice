# 외국인+경제성장률

```python
import pandas as pd
import numpy as np

foreign = pd.read_excel('/content/drive/MyDrive/시계열/외국인 보유량.xlsx')
sox = pd.read_excel('/content/drive/MyDrive/시계열/필라델피아 반도체 지수.xlsx')
USgdp = pd.read_excel('/content/drive/MyDrive/시계열/경제성장률-미국.xlsx')
KRgdp = pd.read_excel('/content/drive/MyDrive/시계열/경제성장률-한국.xlsx')
CNgdp = pd.read_excel('/content/drive/MyDrive/시계열/경제성장률-중국.xlsx')
```

```python
foreign['날짜'] = pd.to_datetime(foreign['날짜'])

foreign = foreign.sort_values(by='날짜', ascending=True)

print(foreign)
```

```python
# 날짜 열에서 한글 형식의 데이터를 표준화
sox['날짜'] = sox['날짜'].str.replace('월', '').str.replace(',', '')

# pd.to_datetime으로 변환
sox['날짜'] = pd.to_datetime(sox['날짜'], format='%m %d %Y')

# 원하는 형식으로 변환 (YYYY-MM-DD)
sox['날짜'] = sox['날짜'].dt.strftime('%Y-%m-%d')

# 변환 결과 확인
sox.head()
```

```python
sox = sox.sort_values(by='날짜', ascending=True)
print(sox)
```

```python
foreign['날짜'] = pd.to_datetime(foreign['날짜'])
sox['날짜'] = pd.to_datetime(sox['날짜'])
merged_df = pd.merge(foreign, sox, on='날짜', how='outer')

# 결과 확인
print(merged_df)
```

```python
USgdp = USgdp.rename(columns={'Date': '날짜'})
USgdp['날짜'] = pd.to_datetime(USgdp['날짜'])
print(USgdp)
```

```python
KRgdp = KRgdp.rename(columns={'Date': '날짜'})
KRgdp['날짜'] = pd.to_datetime(KRgdp['날짜'])
print(KRgdp)
```

```python
CNgdp = CNgdp.rename(columns={'Date': '날짜'})
CNgdp['날짜'] = pd.to_datetime(CNgdp['날짜'])
print(CNgdp)
```

```python
USgdp['날짜'] = pd.to_datetime(USgdp['날짜'])
KRgdp['날짜'] = pd.to_datetime(KRgdp['날짜'])
CNgdp['날짜'] = pd.to_datetime(CNgdp['날짜'])

# 먼저 외국인 데이터와 sox 데이터 합치기
merged_df = pd.merge(foreign, sox, on='날짜', how='outer')

# USgdp, KRgdp, CNgdp 데이터를 순차적으로 merge
merged_df = pd.merge(merged_df, USgdp, on='날짜', how='outer')
merged_df = pd.merge(merged_df, KRgdp, on='날짜', how='outer')
merged_df = pd.merge(merged_df, CNgdp, on='날짜', how='outer')

# 결과 확인
print(merged_df)
```

```python
merged_df['날짜'] = pd.to_datetime(merged_df['날짜']).dt.date
```

```python
merged_df.to_excel('/content/drive/MyDrive/merged_data.xlsx', index=False)
```

# Interpolation

```python
sw = pd.read_excel('/content/drive/MyDrive/시계열/sw_merge.xlsx')
sw.set_index('Date')
sw.head()
```

```python
sw['Date'] = pd.to_datetime(sw['Date'])

# Date 열을 전체 범위로 재생성 (최소~최대 범위 생성, 일 단위)
full_date_range = pd.date_range(start=sw['Date'].min(), end=sw['Date'].max(), freq='D')

# 기존 데이터프레임을 새로운 날짜 범위로 리인덱싱
sw = sw.set_index('Date').reindex(full_date_range)

# 인덱스에서 날짜를 다시 열로 변환
sw.index.name = 'Date'
sw = sw.reset_index()

# 필요한 열에서 선형 보간 수행 (GDP 관련 열)
gdp_columns = ['US Economic Growth Rate', 'KR Economic Growth Rate','US GDP Annual', 'KR GDP Annual', 'CN GDP Annual']
sw[gdp_columns] = sw[gdp_columns].interpolate(method='linear')

sw.head()
```

```python
sw['Date'] = pd.to_datetime(sw['Date']).dt.date
sw.to_excel('/content/drive/MyDrive/interpolation.xlsx', index=False)
```
