```python
# 변환 함수 정의
def convert_date(row):
    date_str = row['mmdd']
    index = row.name  # 현재 행의 인덱스 가져오기

    # 연도 설정: 0~246은 2001, 그 외는 2002
    if index <= 246:
      year = 2001
    elif index <= 496:
      year = 2002
    elif index <= 746:
      year = 2003
    else:
      year = 2004

    # 정규식으로 월과 일을 추출
    match = re.match(r"(\d{1,2})월(\d{1,2})일", date_str)
    if match:
        month, day = match.groups()
        return f"{year}-{int(month)}-{int(day)}"
    return None  # 변환 실패 시 None 반환
```

## 폰 점유율

```python
from scipy.interpolate import splrep, splev
```

```python
phone_market_share = pd.read_excel('/content/drive/MyDrive/Time Series Project/Stock Price/data/폰 점유율.xlsx')
```

```python
phone_market_share.head(20)
```

```python
start_date = phone_market_share['Date'].min()
end_date = pd.Timestamp('2024-09-30')
full_date_range = pd.date_range(start = start_date, end = end_date, freq = 'B')
```

```python
phone_market_share.info()
```

```python
phone_market_share['Date'] = pd.to_datetime(phone_market_share['Date'])
phone_market_share = phone_market_share.set_index('Date')
```

```python
phone_market_share.info()
```

```python
phone_market_mean = phone_market_share.reindex(full_date_range, method = None)
phone_market_mean = phone_market_mean['Samsung'].fillna(method = 'ffill')
phone_market_mean
```

```python
phone_market_mean.info()
```

```python
dates_ordinal = phone_market_share.index.map(pd.Timestamp.toordinal).values
values = phone_market_share['Samsung'].values

# 3. 전체 주중 날짜 범위 생성
start_date = phone_market_share.index.min()
end_date = pd.Timestamp('2024-09-30')
full_date_range = pd.date_range(start=start_date, end=end_date, freq='B')
full_date_range_ordinal = full_date_range.map(pd.Timestamp.toordinal).values

# 4. 스플라인 보간 설정
spl = splrep(dates_ordinal, values)  # Fit spline to the original data
interpolated_values = splev(full_date_range_ordinal, spl)  # Interpolate for the full range

# 5. 결과 데이터프레임 생성
spline_interpolated_data = pd.DataFrame({
    'Date': full_date_range,
    'Samsung_spline': interpolated_values
})

# 결과 확인
spline_interpolated_data.head()
```

```python
spline_interpolated_data['Date'] = pd.to_datetime(spline_interpolated_data['Date'])
spline_interpolated_data = spline_interpolated_data.set_index('Date')
spline_interpolated_data
```
